```python
#"Anji - Restoring the Aatma Lingam" - a game designed by Sree Lakshmi Sravya
"""
    Docstring:
    
    A) This game is based on a Telugu fantasy film, Anji (2004). It consists 
    of three stages and has defined functions for start, win and loose.
    This game requires wisdom, luck and alot of bravery too. Having a tad bit 
    of knowledge about the movie would help. Be the mentor of Anji through his 
    journey to restore the 'Aatma Lingam'.
    
    Stage 1: Answer the question.
    Stage 2: Cross the Hell's bridge.
    Stage 3: Unlock the Sanctum Sanctorum.
    
    B) Known Bugs and/or Errors:
    None.
    
"""

#Import functions
from sys import exit 
import random 

#defining global varialbles
#Introducing the protagonist
hero_name = "Anji"
    
#Introducing the Godfather
godfather_name = "Sivanna"
 
user_name = " " 


# Base for running the game

def run_game():
    #Introducing the background story of the game and movie.
    
    print(f""" 
    {'*' * 35} ANJI- RESTORING THE AATMA LINGAM {'*' * 35}\n
    The tale of Anji & his trials to restore the 'Aatma Lingam', that can grant 
    immortality and supernatural powers, unfolds here. Make wise choices, or 
    you die.  Once you get in, the only way out is death.\n
    { '*' * 105}
    """)
    

    def start_game():
        start_prompt=input(prompt = "\n<Press ENTER to start/Continue the game!> \n")
        if start_prompt == "":
            print("")
        else:
            print("\n you pressed something other than 'ENTER' on keyboard")
            start_game()
    start_game() 



    print(f"""\n\n
        {godfather_name}: Namaste! I am {godfather_name}, {hero_name}'s guardian. 
        \n What is your Name? \n\n""")


    def valid_name():
        global user_name
        user_name = input(prompt = "Please enter a name with all letters as alphabets and with atleast 3 letters")
        if len(user_name) > 2 and user_name.isalpha():
            user_name = user_name.upper()
            print(f""" 
        {godfather_name}: Hi {user_name}!! Thank you for agreeing to help our {hero_name}. 
        I would now leave both the 'Aatma Lingam' & {hero_name} in your care.\n\n""")
            start_game() 
        else:
            print("\n Please enter a Valid Name!!")
            valid_name()
    valid_name()
    
    start()
    
    #starting the game
#########################################################################

#########################################################################

def start():
    print(f"""\n\n 
    As Anji & you start on this journey, you need to enter the 'Kalabhairava Temple',
    a place with divine powers. Legend has it that not a single man has returned 
    once he entered this temple. But without entering it, you can not save the 
    'Aatma Lingam' from Bhatia.\n""")
    
    #Asking for the user's choice to continue
    print(f""" \n{hero_name}: {user_name}, you are my guru and I would follow your
    word, no matter what. In this juncture, what would you like us to do?
    1) Enter the temple.
    2) Wait for the next auspicious day to prepare ourselves better.
    3) Forget this mission. \n
    """)
    
    choice = str(input(prompt = "Hi guru, Please choose an option among the choices 1,2 and 3>  \n")).upper()
  
    if choice in ["1","ENTER","TEMPLE","ENTER THE TEMPLE","ENTER TEMPLE"]:
        
        print(f"""\n\n
        {hero_name}: As you say, {user_name}.\n\n
        {'*' * 3} You have now entered the Temple.\n""")
        
        input(prompt= "\n\n <Press ENTER to continue to stage_1> \n")
        stage_1()
        
    elif choice in ["2","WAIT","AUSPICIOUS","WAIT FOR THE NEXT AUSPICIOUS DAY TO PREPARE OURSELVES BETTER"]:

        
        print(f"""\n\n{'*' * 3}72 days later.\n
        {hero_name}: Now that we came back, what do we do, {user_name}?""")
        
        input(prompt = "\n\n<Press ENTER to continue> \n")
        start()
        
    elif choice in ["3","FORGET THIS MISSION","FORGET","FORGET MISSION"]:
        print(f"""\n
        {hero_name} and you run, Bhatia kills you and captures the 
                  'Aatma Lingam' and uses it for his greed. Game over!\n""")
        fail()
        
    else:
        print(f"""\n\n invalid choice.Please Choose again \n""")
        start()
    
        
########################################################################## 
#STAGE 1 :     Answer the question.  
##########################################################################

def stage_1():
    print('-' * 70)
    print(f"""\n\n 
    As you enter the old, pitch- dark temple, you hear a strange voice echo: \n
    "Oh Humans! Don't take a step further, if you do not wish to die. If you want 
    to move further, you need to rightfully answer two questions!""")
    
    #preparing the question list
    question_list = [ 'What is death?', 
                      'Where is God?',
                      'What do you fear the most?',
                      'Do you believe in God?']
    questions = 2
    question_asked=0
    question_list_copy=[ 'What is death?', 
                      'Where is God?',
                      'What do you fear the most?',
                      'Do you believe in God?']
    while questions > 0:
        question = random.choice(question_list_copy)
        question_asked=question_asked+1
        
        input(prompt = "<Press ENTER for your question> \n")
        
        print(question)
        if question == question_list[0]:
            print("""
            1) Loss of body 
            2) Loss of body and soul  
            3) I don't know \n""")
            answer_1 = input(prompt = "choose> \n")
            
            if answer_1 == "2":
                print("Good, you are one step closer to your destination.\n")
                questions -=1
                if question_asked==2:
                    print(" Well done, walk straight to the next hall for the next challenge.")
                    input(prompt = "Press ENTER to walk.")
                    stage_2()         
            
            else: 
                print("""
                Dumb humans! Rot in this dark hall for years as no one can 
                come to your rescue now!""")
                questions=0
                fail()
                
                
            
        elif question == question_list[1]:
            print ("""
            1) Everywhere 
            2) Nowhere 
            3) I do not know. """)
            answer_2 = input(prompt = "Choose among 1,2&3> \n")
            
            if answer_2 == "1":
                print ("Good, you are one step closer to your destination.\n")
                questions -=1
                if question_asked==2:
                    questions=0
                    print(" Well done, walk straight to the next hall for the next challenge.")
                    input(prompt = "Press ENTER to walk.")
                    stage_2()   
                
            else:
                print("""
                Dumb humans! Rot in this dark hall for years 
                as no one can come to your rescue now!\n """)
                questions=0
                fail()
                
        elif question == question_list[2]:
            print(" You can give any input.")
            answer_3 = input(prompt = "> \n")
            print("\n It is good to know you better.\n")
            questions -=1
            if question_asked==2:
                questions=0
                print(" Well done, walk straight to the next hall for the next challenge.")
                input(prompt = "Press ENTER to walk.")
                stage_2()   
            
        elif question == question_list[3]:
            print ("""
            1) Yes! 
            2) No. 
            3) Maybe, not sure.""")
            
            answer_3 = input(prompt= "choose> \n")
            
            if answer_3 == "1":
                print("""\n 
                It is good to know. Hope the Lord helps you in this journey.\n""")
                if question_asked==2:
                    questions=0
                    print(" Well done, walk straight to the next hall for the next challenge.")
                    input(prompt = "Press ENTER to walk.")
                    stage_2()  
                
                
            elif answer_3 == "2":
                print("""\n 
                It takes courage to truly not believe in something 
                that most of the others do.\n""")
                if question_asked==2:
                    questions=0
                    print(" Well done, walk straight to the next hall for the next challenge.")
                    input(prompt = "Press ENTER to walk.")
                    stage_2()  
                
            else:
                print("\n It is okay to not be sure.\n")
                if question_asked==2:
                    questions=0
                    print(" Well done, walk straight to the next hall for the next challenge.")
                    input(prompt = "Press ENTER to walk.")
                    stage_2()  
                
        question_list_copy.remove(question)   
        
            #Establishing that the user won


    #Moving to stage_2
    
#############################################################################    
#Stage 2 -      Cross the Hell's bridge.               
#############################################################################

def stage_2():
    print({'*' * 90})
    print(f""" {'*' * 3} The next hall.""")
    print(f""" 
    {hero_name}: Wow! We passed the first challenge. I 
    wonder what is in store for us in this challenge.""")
    print(f"""
            {hero_name} and you come into a large room with extravagant lights.
            But as soon as you took your 7th step inside, the floor collapsed into
            raging fire from beneath. The floor was unevenly collapsed, leaving 
            bits of floor here and there. The only way to cross this hall is to jump on 
            the bits of floor. You might fall into the fire while trying to cross.
            \n""")
    print(f""" 
    {hero_name}: OMG!!! What is this? Careful, {user_name}! Do not walk further.""")
    print(f""" 
    {user_name}: Yes! Stay close, {hero_name}! I have never seen something
    like this before. But I am sure there are instructions on how to cross this 
    around. Lets look!\n""")
    print(f"""
    {hero_name} and you look around the place.""")
    
    
    #finding the instructions
    print(f"""{user_name}: Hey! Look, there is something written on this pillar.""")
    input(prompt = "<Press ENTER to access the information>")
    print(f"""
            Now that we have tested your wisdom, it is now time to test your
            chivalry. Infront of you lies, the bridge of Hell. If successful, you guys will 
            be the first to make it alive after this challenge. Avoid stepping on every 
            other step and you will be safe. You can only skip one step, at a time.
            See you on the other side.""")
    
    print(f""" \n
    {hero_name}: Does this mean that we have to hop on alternate steps?\n
               
    {user_name}: Seems like it. There are a total of 10 steps. 
    So if we step on the first step, the next one should be on the 
    third step. I will jump on the first one. \n
                            """ )
    
    #asking user to press ENTER to continue
    input(prompt = "<Press ENTER to jump on to the first step.> \n")
    
    print(f"""{'*' * 3} {user_name} jumps onto the first step. Nothing happens. \n""")
    
    #consverstaion between user and ANJI
    print(f"""
    {user_name}: {hero_name}, it is safe. Do not jump until I jump on the next one, okay?""")
    print(f""" 
    {hero_name}: Okay!""")
    
    #asking input for the steps
    print("Choose your next step \n.") 
    step_2 = input(prompt = "Choose your second step> ")
    if step_2 == "3" or step_2 == "three":
        print("Good! Choose your next step.")
        step_3 = input(prompt ="Choose your third step> \n")
        if step_3 == "5" or step_3=="five":
            print("Good! Choose your next step.")
            step_4 = input(prompt = "Choose your fourth step> \n")
            if step_4 == "7" or step_4=="seven":
                print("Good! Choose your next step.")
                step_5 = input(prompt = "Choose your fifth step> \n")
                if step_5 == "9" or step_5=="nine":
                    print("Good! Choose your next step.")
                    print(f""" 
    {'*' * 3} You are on the nineth step. You look back and see {hero_name} 
    on the seventh step.
    {user_name}: {hero_name}! We are so close to finishing this. 
    Balance youself well and jump on the eleventh step to reach the floor.\n
    {hero_name}: Okay! You be careful too!
    """)
                    step_6 = input(prompt = "Choose your sixth step> \n")
                    if step_6 == "11" or step_6 == "eleven":
                        print("Congratulations! You've cleared the mission!.")
                            #establishing that the user won
                        print(f"""\n\n Both {hero_name} and you have made it alive and go to a lone pillar 
    on the other side.\n
    {hero_name}: "If you are reading this, then it means that you've crossed the bridge 
    of Hell, congratualtions! You are very close to witnessing the greatest sanctorum
    in the whole wide world. Walk forward to reach the sanctorum."\n
    {user_name}: Okay, let's walk ahead now!\n
    """)
                    #Moving to stage 3
    
                        stage_3 ()
                    else:
                        print("You have chosen the wrong step. You fall into the fiery pit and burn.")
                        fail()
                else:
                    print("You have chosen the wrong step. You fall into the fiery pit and burn.")
                    fail()
            else:
                print("You have chosen the wrong step. You fall into the fiery pit and burn.")
                fail()
        else:
            print("You have chosen the wrong step. You fall into the fiery pit and burn.")
            fail()
    else:
        print("You have chosen the wrong step. You fall into the fiery pit and burn.")
        fail()

 
            


#############################################################################
#Stage 3: Unlocking the Sanctum Sanctorum.
#############################################################################

def stage_3():
    print(f"{'*' * 90}")
    print(f"""\n\n 
    {hero_name}: Finally! We are so close to our goal. We can restore the 'Aatma Lingam'
    in its original place and make sure that Bhatia can never lay his filty hands on it again.
                   
    {user_name}: True. This is the door to the Holy Sanctorum. I think this is where the 
    'Aatma Lingam' immerses itself into Lord Shiva's soul. But how do you open this door?
                   
    {hero_name}: Let us try knocking it.
    \n""")
    #asking user to press ENTER to knock
    input(prompt="\n\n<Press ENTER to knock on the door> \n")
          
    # Printing the script on the door
    print(f""" \n\n{'*' * 3}Words form on the door in Sanskrit. \n
          {user_name}:"Congratulations on making it this far. You have shown 
          excellent wisdom and courage. Now it is time to test your luck. 
          You have five chances to guess the right combination of numbers for 
          the password of this door. If you get it right, it opens to your goal, 
          orelse, it opens to a room filled with vicious snakes. Good luck!"
          {hero_name}: What? what does this mean? So if we don't guess the number
          in five chances, are we going to die?
          {user_name}: Most definitely! And that too, not a pretty death!
          {hero_name}: But how do we guess them, {user_name}?
          {user_name}: Just randomly I guess, it is a luck test for a reason.
        \n\n""")
          
    #asking the user to press ENTER to start the game
    input(prompt = "<Press ENTER to start guessing the number.> \n")
    print("""\n\n Young one! Give me a number between 10 and 20, you have five 
    chances to get it right. \n""")
          
    
    #Setting the passwrod between 10 and 20
    real_password = random.randint(10, 20)
          
    #Establishing a loop on the password
    for i in range(5):   
    
        #asking for the user to guess
        print("\n Guess the number for the password (10-20) \n")
        guess_0 = input(prompt = '')
        guess = int(guess_0)

        # Comparing the guess to the real password
        if guess == real_password:
                print(f"""\n
                You have unlocked the door to the holy sanctorum of the Kalabhairava temple!/n
                {hero_name}: Wow! You have made it, {user_name}! We can now 
                finally go into to the Sanctorum, I can not believe this. \n""")
                print(" \n\n***The door opens to the statue of Lord Shiva's statue.\n")
                print(f""" A voice: {hero_name} and {user_name}, you can now go
                ahead and place the 'Aatma Lingam' in the hollow of Lord Shiva's Chest.""")
                print(f""" \n\n {'-' * 30} THE END {'-' * 30}""")
                break

        
        #Creating outputs for various inputs
        elif guess > real_password:
                print ("""Incorrect! The real password is smaller than you think. 
                       Try again.\n""")

        elif guess < real_password:
                print ("""Incorrect! The real password is larger than you think. 
                Try again.\n""")

        else:
                print("Invalid response. Try again.")


        #Establishing a scenario of failure or exceeding of chances
        if i == 4:
                print(f"""\n\n 
                You have used up all your chances but in vain. The real password 
                is "{real_password}". \n
                {'*' * 3} The door opens to a dark room and you hear multiple 
                hissing sounds. Out of no where you see a gaint Python open its 
                mouth towards you.\n""")
                print(f""" \n\n{'*' * 40} GAME OVER! {'*' * 40}""")
                fail()

        
     

      #THE END

def fail():
    print("\n You've failed in your mission to restore the 'Aatma Lingam'.")
    print("\n Would you like to play again? (yes/no)")
    #Looping until the user inputs correctly
    #while True:
        #Choose yes or no
    replay = input(prompt = "> ")
        #converting lower case
    replay = replay.lower()
        
    if replay == "yes":
        print("\n Okay")
        input (prompt = "press ENTER to start again")
        stage_1()
            
    elif replay == "no":
        print(f""" Thanks for playing, {user_name}!""")
            
    else:
        print("Invalid response.")
        fail()

          

   #run the functions     
          
          
#############################################################################
          #START THE GAME#
#############################################################################
          
run_game()
```
